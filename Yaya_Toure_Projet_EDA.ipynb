# Analyse de la distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution de 'tip'
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_clean['tip'], kde=True, bins=20)
plt.title('Distribution de Tip')

plt.subplot(1, 2, 2)
sns.boxplot(y=df_clean['tip'])
plt.title('Boxplot de Tip')
plt.show()

# Distribution de 'total'
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_clean['total'], kde=True, bins=20)
plt.title('Distribution de Total')

plt.subplot(1, 2, 2)
sns.boxplot(y=df_clean['total'])
plt.title('Boxplot de Total')
plt.show()
```

```python
# Définir les bornes
lower_bound = {
    'tip': df_clean['tip'].quantile(0.03),
    'total': df_clean['total'].quantile(0.03)
}

upper_bound = {
    'tip': df_clean['tip'].quantile(0.96),
    'total': df_clean['total'].quantile(0.96)
}

# Filtrer les outliers pour chaque colonne séparément
df_no_outliers = df_clean[
    (df_clean['tip'] >= lower_bound['tip']) & (df_clean['tip'] <= upper_bound['tip']) &
    (df_clean['total'] >= lower_bound['total']) & (df_clean['total'] <= upper_bound['total'])
]

# Afficher la nouvelle taille des données
print(df_no_outliers.shape)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que df_clean soit déjà nettoyé des valeurs manquantes et doublons
# Vous pouvez charger votre jeu de données ici si nécessaire.

# Calcul de l'IQR pour détecter les outliers
Q1 = df_clean[['tip', 'total']].quantile(0.25)
Q3 = df_clean[['tip', 'total']].quantile(0.75)
IQR = Q3 - Q1

# Définir des bornes pour les outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrer les outliers pour chaque colonne séparément
df_no_outliers = df_clean[
    (df_clean['tip'] >= lower_bound['tip']) & (df_clean['tip'] <= upper_bound['tip']) &
    (df_clean['total'] >= lower_bound['total']) & (df_clean['total'] <= upper_bound['total'])
]

# Tracer les boxplots pour 'tip' et 'total' après le nettoyage des outliers
plt.figure(figsize=(10, 5))

# Boxplot pour la colonne 'tip'
plt.subplot(1, 2, 1)
sns.boxplot(y='tip', data=df_no_outliers)
plt.title('Boxplot de Tip (sans valeurs extrêmes)')

# Boxplot pour la colonne 'total'
plt.subplot(1, 2, 2)
sns.boxplot(y='total', data=df_no_outliers)
plt.title('Boxplot de Total (sans valeurs extrêmes)')

# Affichage des graphiques
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
# Appliquer la transformation logarithmique en utilisant .loc pour éviter l'avertissement
df_no_outliers.loc[:, 'log_tip'] = np.log(df_no_outliers['tip'] + 1)  # Ajout de 1 pour éviter log(0)
df_no_outliers.loc[:, 'log_total'] = np.log(df_no_outliers['total'] + 1)

# Tracer les boxplots après la transformation logarithmique
plt.figure(figsize=(10, 5))

# Boxplot pour la colonne 'log_tip'
plt.subplot(1, 2, 1)
sns.boxplot(y='log_tip', data=df_no_outliers)
plt.title('Boxplot de Log(Tip)')

# Boxplot pour la colonne 'log_total'
plt.subplot(1, 2, 2)
sns.boxplot(y='log_total', data=df_no_outliers)
plt.title('Boxplot de Log(Total)')

plt.tight_layout()
plt.show()

# Tracer les distributions après la transformation logarithmique
plt.figure(figsize=(10, 5))

# Distribution pour la colonne 'log_tip'
plt.subplot(1, 2, 1)
sns.histplot(df_no_outliers['log_tip'], kde=True)
plt.title('Distribution de Log(Tip)')

# Distribution pour la colonne 'log_total'
plt.subplot(1, 2, 2)
sns.histplot(df_no_outliers['log_total'], kde=True)
plt.title('Distribution de Log(Total)')

plt.tight_layout()
plt.show()
```

```python
# Recalculer les bornes IQR et appliquer le filtre pour supprimer les outliers dans 'log_total'
Q1_log_total = df_no_outliers['log_total'].quantile(0.25)
Q3_log_total = df_no_outliers['log_total'].quantile(0.75)
IQR_log_total = Q3_log_total - Q1_log_total

# Définir les bornes pour les outliers dans 'log_total'
lower_bound_log_total = Q1_log_total - 1.5 * IQR_log_total
upper_bound_log_total = Q3_log_total + 1.5 * IQR_log_total

# Filtrer les données pour supprimer les outliers dans 'log_total'
df_no_outliers_filtered = df_no_outliers[(df_no_outliers['log_total'] >= lower_bound_log_total) &
                                         (df_no_outliers['log_total'] <= upper_bound_log_total)]

# Tracer le boxplot et la distribution après avoir supprimé les outliers de 'log_total'
plt.figure(figsize=(10, 5))

# Boxplot pour la colonne 'log_total' sans outliers
plt.subplot(1, 2, 1)
sns.boxplot(y='log_total', data=df_no_outliers_filtered)
plt.title('Boxplot de Log(Total) (sans outliers)')

# Distribution pour la colonne 'log_total' sans outliers
plt.subplot(1, 2, 2)
sns.histplot(df_no_outliers_filtered['log_total'], kde=True)
plt.title('Distribution de Log(Total) (sans outliers)')

plt.tight_layout()
plt.show()
```

```python
# Recalculer les bornes IQR pour 'log_tip' et 'log_total' afin de supprimer les outliers
Q1_log_tip = df_no_outliers['log_tip'].quantile(0.25)
Q3_log_tip = df_no_outliers['log_tip'].quantile(0.75)
IQR_log_tip = Q3_log_tip - Q1_log_tip

Q1_log_total = df_no_outliers['log_total'].quantile(0.25)
Q3_log_total = df_no_outliers['log_total'].quantile(0.75)
IQR_log_total = Q3_log_total - Q1_log_total

# Définir les bornes pour les outliers dans 'log_tip' et 'log_total'
lower_bound_log_tip = Q1_log_tip - 1.5 * IQR_log_tip
upper_bound_log_tip = Q3_log_tip + 1.5 * IQR_log_tip

lower_bound_log_total = Q1_log_total - 1.5 * IQR_log_total
upper_bound_log_total = Q3_log_total + 1.5 * IQR_log_total

# Filtrer les données pour supprimer les outliers dans 'log_tip' et 'log_total'
df_no_outliers_filtered = df_no_outliers[
    (df_no_outliers['log_tip'] >= lower_bound_log_tip) &
    (df_no_outliers['log_tip'] <= upper_bound_log_tip) &
    (df_no_outliers['log_total'] >= lower_bound_log_total) &
    (df_no_outliers['log_total'] <= upper_bound_log_total)
]

# Tracer le boxplot et la distribution après avoir supprimé les outliers de 'log_tip' et 'log_total'
plt.figure(figsize=(12, 6))

# Boxplot pour les colonnes 'log_tip' et 'log_total' sans outliers
plt.subplot(2, 2, 1)
sns.boxplot(y='log_tip', data=df_no_outliers_filtered)
plt.title('Boxplot de Log(Tip) (sans outliers)')

plt.subplot(2, 2, 2)
sns.boxplot(y='log_total', data=df_no_outliers_filtered)
plt.title('Boxplot de Log(Total) (sans outliers)')

# Distribution pour les colonnes 'log_tip' et 'log_total' sans outliers
plt.subplot(2, 2, 3)
sns.histplot(df_no_outliers_filtered['log_tip'], kde=True)
plt.title('Distribution de Log(Tip) (sans outliers)')

plt.subplot(2, 2, 4)
sns.histplot(df_no_outliers_filtered['log_total'], kde=True)
plt.title('Distribution de Log(Total) (sans outliers)')

plt.tight_layout()
plt.show()
```
